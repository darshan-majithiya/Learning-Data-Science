```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
def lrCostFunction(theta, X, y, Lambda):
    m=len(y)
    predictions=sigmoid(X @ theta)
    error = (-y * np.log(predictions)) - ((1 - y)*np.log(1-predictions))
    cost = 1/m * sum(error)
    regCost = cost + Lambda/(2*m) * sum(theta[1:]**2)
    
    # compute grdient descent
    j_0 = 1/m * (X.transpose() @ (predictions - y))[0]
    j_1 = 1/m * (X.transpose() @ (predictions - y))[1:] + (Lambda/m)*theta[1:]
    grad = np.vstack((j_0[:, np.newaxis], j_1))
    return regCost[0], grad
```

```python
def gradient_descent(alpha, X, y, theta, num_iters, Lambda):
    m=len(y)
    J_history = []
    for i in range(num_iters):
        cost, grad = lrCostFunction(theta, X, y, Lambda)
        J_history.append(cost)
        theta = theta - (alpha*grad)
    return J_history, theta
```

```python
def oneVsAll(X, y, num_labels, Lambda):
    m,n = X.shape[0], X.shape[1]
    all_J = []
    all_theta = []
    initial_theta = np.zeros((n+1, 1))
    
    X = np.hstack((np.ones((m,1)), X))
    
    for i in range(1, num_labels+1):
        J_history, theta = gradient_descent(1, X, np.where(y==i, 1, 0), initial_theta,300,Lambda)
        all_J.extend(J_history)
        all_theta.extend(theta)
    return np.array(all_theta).reshape(num_labels, n+1), all_J
```

```python
all_theta, all_J = oneVsAll(X, y, 10, 0.1)
```

```python
plt.plot(all_J[0:300])
plt.xlabel('Iterations')
plt.ylabel('$J(\Theta)$')
plt.title('Cost Function using Gradient Descent')
plt.show()
```

```python
def predictOneVsAll(all_theta, X):
    m=X.shape[0]
    X = np.hstack((np.ones((m,1)),X))
    predictions = X @ all_theta.T
    return np.argmax(predictions, axis=1)+1
```

## Getting higher accuracy using neural network
For the first part of the exercise, the optimized theta values were given to us and we are supposed to implement feedforward propagation to obtain the prediction and model accuracy.

```python
data2 = loadmat('ex3weights.mat')
```

```python
theta1 = data2['Theta1']
theta2 = data2['Theta2']
```

```python
theta1.shape, theta2.shape
```

```python
def predict(theta1, theta2, X):
    '''
    Predict the label of an input given a trained Neural Network.
    '''
    
    # input layer
    m = X.shape[0] 
    X = np.hstack((np.ones((m,1)), X))
    
    # hidden layer
    a1 = sigmoid(X @ theta1.T)
    a1 = np.hstack((np.ones((m,1)), a1))
    
    # output layer
    a2 = sigmoid(a1 @ theta2.T)
    return np.argmax(a2, axis=1)+1
```

```python
prediction = predict(theta1, theta2, X)
```

```python
print('Training set accuracy trained using Neural Network: ', sum(prediction[:, np.newaxis]==y)[0]/5000 * 100, '%.')
```
