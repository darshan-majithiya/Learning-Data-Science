```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#weather data
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y, color='g',linewidth=2.0, linestyle=':', marker='+')
plt.ylabel('Temperature')
plt.xlabel('Date')
plt.title('Weather data')
plt.show()
```

```python
plt.plot(x,y, '.r-')
plt.ylabel('Temperature')
plt.xlabel('Date')
plt.title('Weather data')
plt.show()
```

```python
plt.plot(x,y, '.')
plt.ylabel('Temperature')
plt.xlabel('Date')
plt.title('Weather data')
plt.show()
```

```python
plt.plot(x,y, color='g',linewidth=2.0, markersize=10,linestyle=':', marker='+')
plt.ylabel('Temperature')
plt.xlabel('Date')
plt.title('Weather data')
plt.show()
```

```python
plt.plot(x,y, color='g',linewidth=2.0, alpha = 0.2)
plt.ylabel('Temperature')
plt.xlabel('Date')
plt.title('Weather data')
plt.show()
```

```python
plt.plot(x,y, color='g',linewidth=2.0, alpha = 1)
plt.ylabel('Temperature')
plt.xlabel('Date')
plt.title('Weather data')
plt.show()
```

```python
#data
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.plot(days, avg_t, label = 'Avg')
plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Weather Data')
plt.legend()
```

```python
plt.plot(days, avg_t, label = 'Avg')
plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Weather Data')
plt.legend(loc = 'upper right')
```

```python
plt.plot(days, avg_t, label = 'Avg')
plt.plot(days, max_t, label = 'Max')
plt.plot(days, min_t, label = 'Min')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Weather Data')
plt.legend(loc = 'best', shadow = True) # loc = 'best' default
```
