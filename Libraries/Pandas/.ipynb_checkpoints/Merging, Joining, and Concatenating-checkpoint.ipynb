## Merging by Indexes with the left_index and right_index Parameters

```python
sales_w1 = pd.read_csv('Datasets/Restaurant - Week 1 Sales.csv')
sales_w2 = pd.read_csv('Datasets/Restaurant - Week 2 Sales.csv')
foods = pd.read_csv('Datasets/Restaurant - Foods.csv', index_col='Food ID')
customers = pd.read_csv('Datasets/Restaurant - Customers.csv', index_col = 'ID')
```

```python
customers.head()
```

```python
sales_w1.head()
```

```python
sales_w1 = sales_w1.merge(customers, how='left', left_on='Customer ID', right_index=True)
```

```python
sales_w1.head()
```

```python
foods.head()
```

```python
sales_w1=sales_w1.merge(foods, how='left', left_on='Food ID', right_index=True)
```

```python
sales_w1.head()
```

## pd.merge() method

```python
sales_w1 = pd.read_csv('Datasets/Restaurant - Week 1 Sales.csv')
sales_w2 = pd.read_csv('Datasets/Restaurant - Week 2 Sales.csv')
foods = pd.read_csv('Datasets/Restaurant - Foods.csv')
customers = pd.read_csv('Datasets/Restaurant - Customers.csv')
satisfaction_w1 = pd.read_csv('Datasets/Restaurant - Week 1 Satisfaction.csv')
```
