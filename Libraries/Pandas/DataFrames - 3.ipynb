## .apply() on the single column

```python
df = pd.read_csv('Datasets/jamesbond.csv', index_col = 'Film')
df.sort_index(inplace = True)
df.head(3)
```

```python
df['Box Office'] = df['Box Office'].apply(lambda x: str(x) + ' millions')
```

```python
df.head()
```

## .apply() method with row values

```python
df.head()
```

```python
def review_movie(row):
    actor = row['Actor']
    budget = row['Budget']
    
    if actor == 'Pierce Brosnan':
        return 'The Best'
    elif actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    else:
        return 'I have no clue'
```

```python
df['review'] = df.apply(review_movie, axis = 1)
```

```python
df
```
