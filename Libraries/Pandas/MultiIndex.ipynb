## .pivot() method

```python
sales = pd.read_csv('Datasets/salesmen.csv', parse_dates = ['Date'])
sales.head(3)
```

```python
sales.info()
```

```python
sales.Salesman = sales.Salesman.astype('category')
```

```python
sales.info()
```

```python
sales = sales.pivot(index = 'Date', columns = 'Salesman', values = 'Revenue')
```

```python
sales.head()
```

```python
sales.info()
```

## .pivot_table() method

```python
food = pd.read_csv('Datasets/foods.csv')
food.head()
```

```python
food.info()
```

```python
food.pivot_table(values = 'Spend', index = ['Gender','Item'], columns = ['City','Frequency'], aggfunc = 'count')
```
