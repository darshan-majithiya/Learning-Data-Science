```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
# ideal equation: y = 2x + 3
```

```python
plt.scatter(x,y,color='red',marker='+',linewidth='5')
```

```python
def gradient_descent(x,y):
    n = len(x)
    theta_a = theta_b = 0
    alpha = 0.00622
    iterations = 4000
    plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(iterations):
        y_predicted = theta_a + theta_b*x
        plt.plot(x, y_predicted, color='g')
        cost = (1/(2*n))*sum((y_predicted-y)**2)
        print('iteration: {}, theta_a: {}, theta_b: {}, cost: {}'.format(i, theta_a, theta_b, cost))
        theta_a_d = (1/n)*sum((y_predicted - y))
        theta_b_d = (1/n)*sum((y_predicted - y)*x)
        theta_a = theta_a - alpha*theta_a_d
        theta_b = theta_b - alpha*theta_b_d
        
    return theta_a, theta_b
```

```python
theta_a, theta_b = gradient_descent(x,y)
```

```python
output = lambda x: theta_b*x + theta_a
```
